```python
import pandas as pd
```

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
```

```python
drinks.head()
```

#### 11.1 How to remove a column  (for example - continent) ?

```python
drinks.drop('continent',axis = 1).head(5)
```

Continent column is dropped , as (inplace = True ) is not used this doesnt change the underlying data. Here axis = 1 states that mentioned string is a column name. Below is an example of dropping rows

#### 11.2 How to remove a row  ?

```python
drinks.drop(0,axis = 0).head(5)
```

row 0 i.e Afghanistan is dropped.

###### axis = 1 to use while dropping column 
###### axis = 0 to use while dropping row

#### 11.3 Using axis with mathematical operator ?

```python
drinks.mean()
```

Here pandas is providing  mean by considering rows for a particular column i.e by default its using axis=0 what if user tries to get mean for all the columns at particular row. For example if i want to look at mean for Albania for (beer_servings	spirit_servings	wine_servings	total_litres_of_pure_alcohol)

```python
drinks.mean(axis=1).head()
```

Above we can see now we have column level mean i.e left to right comining all column.

```python
drinks.mean(axis=1).shape
```

```python
drinks.mean(axis=0).shape
```

Above two blocks explain how axes is working in both the scenario.
