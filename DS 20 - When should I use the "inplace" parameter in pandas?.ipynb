## When should I use the "inplace" parameter in pandas?

```python
import pandas as pd
```

```python
ufo=pd.read_csv('http://bit.ly/uforeports')
```

```python
ufo.head()
```

```python
ufo.shape
```

```python
ufo.drop('City', axis=1).head()
```

drop method is used in the above code to remove City column from ufo dataframe. Next we will check if it is actually dropped.

```python
ufo.head()
```

city column is still present in dataframe.The reason is by default drop column has inplace=False that means that it will not effect the underlying data and can be checked using shift+tab on 'City' used in drop method. 

However this can be altered by using inplace=true. Below is an example.

```python
ufo.drop('City', axis=1, inplace=True)
```

```python
ufo.head()
```

```python
ufo.dropna(how='any').head()
```
